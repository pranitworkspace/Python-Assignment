1. What is the result of the code, and explain?
```python
X = 'iNeuron'
def func():
    print(X)

func()
```
Output:
'iNeuron'

Explanation:
The code defines a global variable `X` with the value 'iNeuron'. Then, a function `func()` is defined that prints the value of `X`. When the function `func()` is called, it accesses the global variable `X` and prints its value, which is 'iNeuron'.

2. What is the result of the code, and explain?
```python
X = 'iNeuron'
def func():
    X = 'NI!'

func()
print(X)
```
Output:
'iNeuron'

Explanation:
The code defines a global variable `X` with the value 'iNeuron'. Then, a function `func()` is defined that assigns a new value 'NI!' to a local variable `X`. However, the function is called but does not print or return anything. After the function call, `print(X)` is executed, which accesses the global variable `X` and prints its value, which remains 'iNeuron' as the local assignment inside the function does not affect the global variable.

4. What output does this code produce? Why?
```python
X = 'iNeuron'
def func():
    global X
    X = 'NI'

func()
print(X)
```
Output:
'NI'

Explanation:
The code defines a global variable `X` with the value 'iNeuron'. Inside the function `func()`, the `global` keyword is used to declare that `X` refers to the global variable, not a local variable. Then, the value of `X` is changed to 'NI'. When the function is called, it modifies the global variable `X` to 'NI'. After the function call, `print(X)` is executed, which accesses the modified global variable and prints its value, which is 'NI'.

6. How about this code: what is its output in Python 3, and explain?
```python
def func():
    X = 'NI'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)

func()
```
Output:
'Spam'

Explanation:
The code defines a function `func()` that includes a nested function `nested()`. Inside `func()`, a local variable `X` is assigned the value 'NI'. The `nonlocal` keyword is used in `nested()` to declare that `X` refers to the variable in the nearest enclosing scope, which is `func()`. Then, the value of `X` is changed to 'Spam'. When `func()` is called, it executes `nested()`, which modifies the value of `X`. Finally, `print(X)` is executed inside `func()`, which prints the modified value 'Spam'.
