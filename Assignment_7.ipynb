# Solution:-

1. What is the name of the feature responsible for generating Regex objects?

feature which is  responsible for generating Regex objects is the re module in Python.

2. Why do raw strings often appear in Regex objects?

In the Raw strings are often used in Regex objects to avoid any unwanted interpretation of special characters. Raw strings treat backslashes as literal characters.

3. What is the return value of the search() method?

The search() method returns a Match object if a match is found in the string, otherwise it returns None.

4. From a Match item, how do you get the actual strings that match the pattern?

for the actual strings that match the pattern from a Match object, you can use the group() method. 

5. In the regex created from the r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group zero cover? Group 2? Group 1?

Group 0 covers the entire matched string.
Group 1 covers the first set of three digits surrounded by parentheses.
Group 2 covers the second set of three digits followed by a hyphen and four digits.

6. In standard expression syntax, parentheses and intervals have distinct meanings. How can you tell a regex that you want it to fit real parentheses and periods?

To indicate that you want to match literal parentheses and periods in a regex pattern, you need to escape them using a backslash (\). 

7. The findall() method returns a string list or a list of string tuples. What causes it to return one of the two options?

The findall() method returns a list of strings if the regex pattern does not contain any capturing groups ( ). If the pattern does contain capturing groups, it returns a list of string tuples.

8. In standard expressions, what does the | character mean?

The | character represents the OR operator. It allows you to specify multiple alternatives for a pattern.

9. In regular expressions, what does the . character stand for?

The . character (dot) is a special metacharacter that matches any character except a newline character. 

10. In regular expressions, what is the difference between the + and * characters?

The + character matches one or more occurrences of the preceding element. It requires the element to appear at least once.the * character matches zero or more occurrences of the preceding element. 

11. What is the difference between {4} and {4,5} in a regular expression?

{4} tells preceding element must occur exactly 4 times.
{4,5} tells preceding element must occur at least 4 times and at most 5 times.

12. What do the \d, \w, and \s shorthand character classes signify in regular expressions?

In regular expressions, \d represents any digit character (0-9).
\w represents any alphanumeric character (a-z, A-Z, 0-9, and underscore).
\s represents any whitespace character (spaces, tabs, newlines, etc.).

13. What do the \D, \W, and \S shorthand character classes signify in regular expressions?

\D represents any non-digit character.
\W represents any non-alphanumeric character.
\S represents any non-whitespace character.

14. What is the difference between .*? and .*?

.*? is a non-greedy match that matches as few characters as possible.
.* is a greedy match that matches as many characters as possible.

15. What is the syntax for matching both numbers and lowercase letters with a character class?

The syntax for matching both numbers and lowercase letters in a character class is [0-9a-z].

16. What is the procedure for making a regular expression case insensitive?

To make a regular expression case insensitive, you can pass the re.IGNORECASE flag as second argument to re.compile() function.

17. What does the . character normally match? What does it match if re.DOTALL is passed as the second argument in re.compile()?

The . (dot) character normally matches any character except a newline character. If re.DOTALL is passed as the second argument in re.compile(), the . character will match any character.

18. If numReg = re.compile(r'\d+'), what will numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen') return?

The numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen') will return the string 'X drummers, X pipers, five rings, X hen'.

19. What does passing re.VERBOSE as the second argument to re.compile() allow you to do?

Passing re.VERBOSE as the second argument to re.compile() allows you to include whitespace and comments within the regular expression pattern for better readability.

20. How would you write a regex that matches a number with a comma for every three digits? It must match the following:

Pattern: r'^\d{1,3}(,\d{3})*$'
Matches: '42', '1,234', '6,368,745'
Does not match: '12,34,567' (only two digits between commas), '1234' (lacks commas)

22. How would you write a regex that matches the full name of someone whose last name is Watanabe? You can assume that the first name that comes before it will always be one word that begins with a capital letter. The regex must match the following:

Regex pattern: r'^[A-Z][a-zA-Z]*\sWatanabe$'
Matches: 'Haruto Watanabe', 'Alice Watanabe', 'RoboCop Watanabe'
Does not match: 'haruto Watanabe' (where the first name is not capitalized), 'Mr. Watanabe' (where the preceding word has a non-letter character), 'Watanabe' (which has no first name), 'Haruto watanabe' (where Watanabe is not capitalized)
