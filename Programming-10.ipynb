1. Write a Python program to find the sum of elements in a list?

```python
def sum_of_elements(lst):
    return sum(lst)


my_list = [1, 2, 3, 4, 5]
print(sum_of_elements(my_list))  # Output: 15
```

2. Write a Python program to multiply all numbers in a list?

```python
def multiply_numbers(lst):
    result = 1
    for num in lst:
        result *= num
    return result


my_list = [1, 2, 3, 4, 5]
print(multiply_numbers(my_list))  # Output: 120
```

3. Write a Python program to find the smallest number in a list?

```python
def find_smallest(lst):
    return min(lst)


my_list = [5, 2, 9, 1, 7]
print(find_smallest(my_list))  # Output: 1
```

4. Write a Python program to find the largest number in a list?

```python
def find_largest(lst):
    return max(lst)


my_list = [5, 2, 9, 1, 7]
print(find_largest(my_list))  # Output: 9
```

5. Write a Python program to find the second largest number in a list?

```python
def find_second_largest(lst):
    sorted_lst = sorted(lst)
    return sorted_lst[-2]


my_list = [5, 2, 9, 1, 7]
print(find_second_largest(my_list))  # Output: 7
```

6. Write a Python program to find the N largest elements from a list?

```python
def find_n_largest(lst, n):
    sorted_lst = sorted(lst, reverse=True)
    return sorted_lst[:n]


my_list = [5, 2, 9, 1, 7]
print(find_n_largest(my_list, 3))  # Output: [9, 7, 5]
```

7. Write a Python program to print even numbers in a list?

```python
def print_even_numbers(lst):
    for num in lst:
        if num % 2 == 0:
            print(num)


my_list = [1, 2, 3, 4, 5]
print_even_numbers(my_list)  # Output: 2, 4
```

8. Write a Python program to print odd numbers in a list?

```python
def print_odd_numbers(lst):
    for num in lst:
        if num % 2 != 0:
            print(num)


my_list = [1, 2, 3, 4, 5]
print_odd_numbers(my_list)  # Output: 1, 3, 5
```

9. Write a Python program to remove empty lists from a list?

```python
def remove_empty_lists(lst):
    return [sublist for sublist in lst if sublist]


my_list = [1, 2, [], 3, [], 4, 5, []]
print(remove_empty_lists(my_list))  # Output: [1, 2, 3, 4, 5]
```

10. Write a Python program to clone or copy a list?

```python
def clone_list(lst):
    return lst.copy()  # or list(lst) or lst[:]


my_list = [1, 2, 3, 4, 5]
cloned_list = clone_list(my_list)
print(cloned_list)  # Output: [1, 2, 3, 4, 5]
```

11. Write a Python program to count the occurrences of an element in a list?

```python
def count_occurrences(lst, element):
    return lst.count(element)


my_list = [1, 2, 2, 3, 2, 4, 2, 5]
print(count_occurrences(my_list, 2))  # Output: 4
```
