1. What exactly is []?

The empty square brackets represent an empty list in Python.

2. In a list of values stored in a variable called spam, how would you assign the value 'hello' as the third value? (Assume [2, 4, 6, 8, 10] are in spam.)

We can assign the value 'hello' to the third index of the list spam using the following code:
spam[2] = 'hello'

3. What is the value of spam[int(int('3' * 2) / 11)]?

The value of spam[int(int('3' * 2) / 11)] is 'd'. This is because '3' * 2 evaluates to '33', which is then converted to the integer 33. 33 divided by 11 equals 3, and so the expression is equivalent to spam[3], which returns the fourth item in the list 'd'.

4. What is the value of spam[-1]?

The value of spam[-1] is the last item in the list 'd'.

5. What is the value of spam[:2]?

The value of spam[:2] is a new list containing the first two items in the list, i.e., ['a', 'b'].

6. What is the value of bacon.index('cat')?
 
The value of bacon.index('cat') is 1. This is because 'cat' appears at index 1 in the list.

7. ow does bacon.append(99) change the look of the list value in bacon?

The append() method adds the value 99 to the end of the list 'bacon'. After the call to bacon.append(99), the list 'bacon' will look like this: [3.14, 'cat', 11, 'cat', True, 99].

8. How does bacon.remove('cat') change the look of the list in bacon?

The remove() method removes the first occurrence of 'cat' from the list 'bacon'. After the call to bacon.remove('cat'), the list 'bacon' will look like this: [3.14, 11, 'cat', True].

9. What are the list concatenation and list replication operators?

The list concatenation operator is the plus symbol (+), and it is used to combine two lists into a new list. The list replication operator is the asterisk symbol (*), and it is used to create a new list by repeating an existing list a specified number of times.

10. What is difference between the list methods append() and insert()?

The append() method adds an item to the end of a list, whereas the insert() method adds an item to a specified index in the list.

11. What are the two methods for removing items from a list?

The two methods for removing items from a list are remove() and pop(). The remove() method removes the first occurrence of a specified item from the list, while the pop() method removes the item at a specified index and returns it.

12. Describe how list values and string values are identical.

List values and string values are both sequences of items in Python. They both support indexing, slicing, and various methods for manipulating and working with them.

13. What's the difference between tuples and lists?

Tuples are similar to lists in that they are sequences of items in Python, but they are immutable, meaning that they cannot be changed once created.

14. How do you type a tuple value that only contains the integer 42?

To create a tuple with a single integer value of 42, we need to add a comma after the integer. The syntax is as follows:
my_tuple = (42,)

15. How do you get a list value's tuple form? How do you get a tuple value's list form?

To get a list value's tuple form, we can use the tuple() function. For example:
my_list = [1, 2, 3]
my_tuple = tuple(my_list)

16. Variables that "contain" list values are not necessarily lists themselves. Instead, what do they contain?

Variables that "contain" list values are actually references to the list object in memory. When we assign a list to a variable, the variable contains a reference to the list, rather than the actual list values.

17. How do you distinguish between copy.copy() and copy.deepcopy()?

copy.copy() creates a shallow copy of an object, meaning that it creates a new object with a new memory address, but the contents of the object are still references to the same objects in memory as the original object. copy.deepcopy() creates a deep copy of an object, meaning that it creates a completely new object with new memory addresses for all of its contents, This means that changes to the contents of the copied object will not affect the original object.
